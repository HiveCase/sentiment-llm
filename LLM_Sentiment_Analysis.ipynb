```python
import json
def llm_sentiment(review):
  response = requests.post('http://aiproxy.sanand.workers.dev/openai/v1/chat/completions',
                           headers = {"Authorization":f"Bearer {api_key}"},
                           json = {
                               "model":"gpt-4o-mini",
                               "messages":[
                                   {"role":"system","content":"Identify the sentiment of the movie. JUST say positive / negative"},
                                   {"role":"user","content":review}
                               ]
                           })
  result = response.json()
  ans = result['choices'][0]['message']['content']
  return ans
```

```python
review_df['llm_sentiment'] = review_df['sentiment'].apply(llm_sentiment)
```
